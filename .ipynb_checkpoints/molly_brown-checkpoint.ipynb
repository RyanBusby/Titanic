```python
def get_df(fname, test=False):
    df = pd.read_csv(fname)

    # create features
    df['cabin_class'] = df.Cabin.apply(cabin_mapping)
    df['num_rooms'] = df.Cabin.apply(num_rooms)

    df['fam_size'] = df.SibSp + df.Parch
    df['nicnamed'] = df.Name.apply(lambda x: 1 if '"' in x else 0)\

    df['est_age'] = df.Age.apply(lambda x: 0 if x.is_integer() else 1)
    df['no_age'] = df.Age.apply(lambda x: 1 if pd.isnull(x) else 0)

    df.Fare = df.Fare.fillna(df.Fare.mean())
    df.Age = df.Age.fillna(df.Age.mean())
    df.Age = np.log(df.Age)
    df.Fare = np.log(df.Fare)

    # df['_FarexRoomsxClass'] = df.Fare+1 * df.num_rooms+1 * df.Pclass

    # encode categorical variables
    # pd.get_dummies(data=df, columns=['A', 'B'])
    df = df.join(
        pd.get_dummies(
            df.Embarked, drop_first=True
        ).add_prefix('embark_')
    )
    df = df.join(pd.get_dummies(df.Sex, drop_first=True))
    df = df.join(
        pd.get_dummies(
            df.cabin_class, drop_first=True
        ).add_prefix('cabin_')
    )
#     df = df.join(pd.get_dummies(df.Pclass).add_prefix('class_'))

#     df['num_rooms_or_x'] = df.num_rooms + df.cabin_X
    # df['Fare_Class'] = df.Fare / df.Pclass

#     df['class1_fare'] = df.Fare * df.class_1
#     df['class2_fare'] = df.Fare * df.class_2
#     df['class3_fare'] = df.Fare * df.class_3

    # df['class3_fare_fam'] = df.class3_fare * df.fam_size
    # drop everything that was encoded or turned into another feature
    drop_list = [
#         'Pclass',
#         'Fare',
#         'num_rooms',
        'Name',
        'Sex',
        'Ticket',
        'Cabin',
        'Embarked',
        'cabin_class',
        'PassengerId',
        'SibSp',
        'Parch',
#         'cabin_X',
#         'class_1',
#         'class_2',
#         'class_3'
    ]
    if test:
        drop_list.remove('PassengerId')
        # drop_list.extend( # these were least important features
        #     [
        #         'no_age',
        #         'embark_Q',
        #         'cabin_C',
        #         'cabin_D',
        #         'cabin_F',
        #         'cabin_G',
        # #         'cabin_T'
        #     ]
        # )
    else:
        drop_list.extend( # test data doesn't have any cabin_T
            [
                # 'no_age',
                # 'embark_Q',
                # 'cabin_C',
                # 'cabin_D',
                # 'cabin_F',
                # 'cabin_G',
                'cabin_T'
            ]
        )
    df = df.drop(drop_list, axis=1).astype(float)

    return df
```

```python
def cabin_mapping(x):
    if pd.isna(x):
        return 'X'
    cabin_class = re.findall(r'[A-Z]', x)[0]
    return cabin_class

def num_rooms(x):
    if pd.isna(x):
        return 0
    num_rooms = len(x.split(' '))
    return num_rooms

def scale_feats(df, cat_vars, train=True, ss=None):
    if train:
        cat_df = df.loc[:,cat_vars].drop('Survived', axis=1)
        ss = StandardScaler()
        scale_df = df.drop(cat_vars, axis=1)
        # x = ss.fit_transform(scale_df.values)
        x = ss.fit_transform(df.drop('Survived', axis=1).values)
        # x = np.hstack([x, cat_df.values])
        return x, ss
    else:
        cat_df = df.loc[:,cat_vars].drop('PassengerId', axis=1)
        # scale_df = df.drop(cat_vars, axis=1)
        # x = ss.transform(scale_df.values)
        x = ss.transform(df.drop('PassengerId', axis=1).values)
        # x = np.hstack([x, cat_df.values])
        return x
```
